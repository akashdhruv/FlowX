```python
# Sample API
import flowx
```

```python
# Define grid parameters
nx, ny = 400, 400
xmin, xmax = -1, 1
ymin, ymax = -1, 1

# Define cell-centered variable names
center_vars   = ['ibmf']
imbound_vars  = center_vars

imbound_info = dict(with_ib = True, mapping_type = 'shapely', verbose = True,
                    ntrees = 20, nquery_trees = 2, nquery_trace = 30)

particle_info = [dict(input='HDF5', file='sm_tshape.h5')]

# Create the grid and data
gridc, gridx, gridy, scalars, particles = flowx.domain.Domain(nx, ny, xmin, xmax, ymin, ymax,
                                              center_vars, particle_info=particle_info)

domain_data_struct = [gridc, gridx, gridy, scalars, particles]

imbound = flowx.imbound.ImBound(domain_data_struct, imbound_vars, imbound_info=imbound_info)

for particle in particles:
    print("Number of points: ", particle.nnp-1)
```

```python
for particle in particles:
    particle.rotate(45)

imbound.map_to_grid()

flowx.io.plot_contour(gridc, 'ibmf', particle=particles[0])
```

```python
for particle in particles:
    particle.rotate(45)
    
imbound.map_to_grid()

flowx.io.plot_contour(gridc, 'ibmf', particle=particles[0])
```

```python
for particle in particles:
    particle.reset()
    
imbound.map_to_grid()

flowx.io.plot_contour(gridc, 'ibmf', particle=particles[0])
```

```python
for particle in particles:
    particle.offset([0.,0.1])
    
imbound.map_to_grid()

flowx.io.plot_contour(gridc, 'ibmf', particle=particles[0])
```
