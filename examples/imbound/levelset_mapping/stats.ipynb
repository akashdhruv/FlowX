```python
import numpy
import matplotlib.pyplot as pyplot

data_crown = numpy.array([773046, 30866364])
data_tshape = numpy.array([706314, 12281904])
data_ref = numpy.array([7e5, 4e7])

ncells = numpy.linspace(4**3,600**3,400)
nqubits = numpy.log2(ncells)
```

```python
pyplot.rc('font', family='serif', size=16)
pyplot.figure()
pyplot.xlabel('ANN Search        Classical Search')
pyplot.ylabel('C')
pyplot.plot(numpy.log10(data_ref), numpy.log10(data_crown))
pyplot.plot(numpy.log10(data_ref), numpy.log10(data_tshape))
pyplot.legend(['CROWN', 'T'])
pyplot.axis('scaled')
pyplot.axis('equal')
pyplot.ylim(5, 8)
pyplot.show()
```

```python
pyplot.rc('font', family='serif', size=16)
pyplot.figure()
pyplot.xlabel(r'$N_{cells}$')
pyplot.ylabel(r'$N_{qubits}$')
pyplot.plot(ncells, nqubits)
pyplot.show()
```
